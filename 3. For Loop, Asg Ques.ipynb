```python
# 12. Create a program that prints the prime numbers in a given range using a `for` loop.

def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def print_primes_in_range(start, end):
    if start < 2:
        start = 2  # Ensure we start at the smallest prime number
    for num in range(start, end + 1):
        if is_prime(num):
            print(num)

# Input the range from the user
start = int(input("Enter the start of the range: "))
end = int(input("Enter the end of the range: "))

print("Prime numbers in the range", start, "to", end, "are:")
print_primes_in_range(start, end)
```

```python
# 14. Write a program to find the maximum element in a 2D list using a nested `for` loop.

def find_max_2d(lst):
    if not lst:
        return None  # Return None if the list is empty
    
    max_element = lst[0][0]  # Initialize max_element with the first element of the list
    
    for row in lst:
        for element in row:
            if element > max_element:
                max_element = element
                
    return max_element

# Example 2D list
matrix = [
    [3, 7, 1],
    [5, 2, 9],
    [8, 4, 6]
]

# Call the function to find the maximum element
max_value = find_max_2d(matrix)

# Display the result
print("The maximum element in the 2D list is:", max_value)
```

```python
# 17. Write a program that converts a list of Fahrenheit temperatures to Celsius using a `for` loop

# Function to convert Fahrenheit to Celsius
def fahrenheit_to_celsius(fahrenheit):
    celsius = (fahrenheit - 32) * 5/9
    return celsius

# List of Fahrenheit temperatures
fahrenheit_temperatures = [32, 68, 86, 104, 122]

# Initialize an empty list to store the Celsius temperatures
celsius_temperatures = []

# Convert Fahrenheit temperatures to Celsius using a for loop
for fahrenheit in fahrenheit_temperatures:
    celsius = fahrenheit_to_celsius(fahrenheit)
    celsius_temperatures.append(celsius)

# Display the converted temperatures
for i in range(len(fahrenheit_temperatures)):
    print(f"{fahrenheit_temperatures[i]}°F = {celsius_temperatures[i]:.2f}°C")
```

```python
# 18. Create a program to print the common elements from two lists using a `for` loop.

# Function to find common elements in two lists
def find_common_elements(list1, list2):
    common_elements = []
    for item1 in list1:
        for item2 in list2:
            if item1 == item2:
                common_elements.append(item1)
    return common_elements

# Example lists
list1 = [1, 2, 3, 4, 5]
list2 = [3, 4, 5, 6, 7]

# Call the function to find common elements
common_elements = find_common_elements(list1, list2)

# Display the common elements
if common_elements:
    print("Common elements in the two lists:", common_elements)
else:
    print("There are no common elements in the two lists.")
```

```python
# 20. Write a program to find the greatest common divisor (GCD) of two numbers using a `for` loop.

# Function to find the GCD of two numbers
def find_gcd(a, b):
    while b:
        a, b = b, a % b
    return a

# Input two numbers from the user
num1 = 8
num2 = 88

# Call the function to find the GCD
gcd = find_gcd(num1, num2)

# Display the GCD
print(f"The GCD of {num1} and {num2} is {gcd}")
```

```python
# 22. Write a program to find the prime factors of a given number using a `for` loop and list comprehension

# Function to find prime factors of a number
def prime_factors(n):
    factors = []
    # Find all the 2s that divide n
    while n % 2 == 0:
        factors.append(2)
        n = n // 2
    # n is now odd
    for i in range(3, int(n**0.5) + 1, 2):
        while n % i == 0:
            factors.append(i)
            n = n // i
    # If n is still greater than 1, it's a prime number itself
    if n > 1:
        factors.append(n)
    return factors

# Input a number from the user
number = int(input("Enter a number to find its prime factors: "))

# Call the function to find prime factors
factors = prime_factors(number)

# Display the prime factors
if factors:
    print(f"The prime factors of {number} are:", factors)
else:
    print(f"{number} is a prime number with no prime factors other than itself.")
```

```python
# 31. Write a program to generate a list of prime numbers using the Sieve of Eratosthenes algorithm and list comprehension.

def sieve_of_eratosthenes(limit):
    sieve = [True] * (limit + 1)
    sieve[0] = sieve[1] = False

    for current in range(2, int(limit**0.5) + 1):
        if sieve[current]:
            for multiple in range(current * current, limit + 1, current):
                sieve[multiple] = False

    return [num for num in range(2, limit + 1) if sieve[num]]

# Input the upper limit from the user
limit = int(input("Enter the upper limit to generate prime numbers: "))

# Call the sieve_of_eratosthenes function to generate prime numbers
prime_numbers = sieve_of_eratosthenes(limit)

# Display the list of prime numbers
print("Prime numbers up to", limit, "are:", prime_numbers)

```

```python
# 32. Create a program that generates a list of all Pythagorean triplets up to a specified limit using list comprehension

# Function to generate Pythagorean triplets up to a specified limit
def generate_pythagorean_triplets(limit):
    triplets = [(a, b, c) for a in range(1, limit + 1)
                          for b in range(a, limit + 1)
                          for c in range(b, limit + 1)
                          if a**2 + b**2 == c**2]
    return triplets

# Input the upper limit from the user
limit = int(input("Enter the upper limit to generate Pythagorean triplets: "))

# Call the function to generate Pythagorean triplets
pythagorean_triplets = generate_pythagorean_triplets(limit)

# Display the list of Pythagorean triplets
print("Pythagorean Triplets up to", limit, "are:")
for triplet in pythagorean_triplets:
    print(triplet)
```

```python
# 34. Write a program that calculates the mean, median, and mode of a list of numbers using list comprehension.

# Example list of numbers
numbers = [2, 3, 2, 5, 6, 3, 4, 5, 2, 3, 4, 5, 6, 7]

# Calculate the mean using list comprehension
mean_value = sum(numbers) / len(numbers)

# Calculate the median using list comprehension
sorted_numbers = sorted(numbers)
if len(numbers) % 2 == 0:
    median_value = (sorted_numbers[len(numbers) // 2 - 1] + sorted_numbers[len(numbers) // 2]) / 2
else:
    median_value = sorted_numbers[len(numbers) // 2]

# Calculate the mode using list comprehension
count_dict = {num: numbers.count(num) for num in set(numbers)}
mode_value = [key for key, value in count_dict.items() if value == max(count_dict.values())]

# Display the results
print("List of Numbers:", numbers)
print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)


```

```python
# 35. Create a program that generates Pascal's triangle up to a specified number of rows using list comprehension.

# Function to generate Pascal's triangle up to a specified number of rows using recursion and list comprehension
def generate_pascals_triangle(num_rows):
    if num_rows == 1:
        return [[1]]
    else:
        previous_triangle = generate_pascals_triangle(num_rows - 1)
        last_row = previous_triangle[-1]
        new_row = [1] + [last_row[i - 1] + last_row[i] for i in range(1, num_rows - 1)] + [1]
        return previous_triangle + [new_row]

# Input the number of rows from the user
num_rows = int(input("Enter the number of rows for Pascal's triangle: "))

# Call the function to generate Pascal's triangle
pascals_triangle = generate_pascals_triangle(num_rows)

# Display Pascal's triangle
print("Pascal's Triangle:")
for row in pascals_triangle:
    print(" ".join(map(str, row)))
```

```python
# 40. Write a program that generates a list of prime palindromic numbers using list comprehension.

# Function to check if a number is prime
def is_prime(num):
    if num <= 1:
        return False
    if num == 2:
        return True
    if num % 2 == 0:
        return False
    for i in range(3, int(num**0.5) + 1, 2):
        if num % i == 0:
            return False
    return True

# Function to check if a number is palindromic
def is_palindromic(num):
    num_str = str(num)
    return num_str == num_str[::-1]

# Use list comprehension to generate prime palindromic numbers
prime_palindromic_numbers = [num for num in range(1, 1000) if is_prime(num) and is_palindromic(num)]

# Display the list of prime palindromic numbers
print("Prime Palindromic Numbers up to 1000:", prime_palindromic_numbers)
```
